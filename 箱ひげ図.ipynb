```python
!apt-get -y install fonts-ipafont-gothic
!pip install japanize-matplotlib
import japanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ファイル名のリスト
file_names = ['read_bytel1t1.csv', 'read_bytel1t2.csv', 'read_bytel1t3.csv', 'read_bytel1t5.csv', 'read_bytel1t10.csv', 'read_bytel1t21.csv']

# 読み取る行番号のリスト
N_values = [-1, 1]

# データを取得してプロットを作成する関数
def plot_data(N, ylabel, title):
    all_data = []

    # 各ファイルからN行目のデータを読み取る
    for file_name in file_names:
        data = pd.read_csv(file_name, header=None)
        if abs(N) <= len(data):
            all_data.append(data.iloc[N].values)
        else:
            print(f"{file_name} does not have {abs(N)} rows.")

    # データが不足している場合は警告を表示
    if len(all_data) < len(file_names):
        print("Some files did not have enough rows. Box plot may not have all expected data.")

    # 図を作成
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=all_data)
    # plt.title(title)
    plt.xlabel('LOOP_COUNT')
    plt.ylabel(ylabel)
    plt.xticks(range(len(file_names)), [1, 2, 3, 5, 10, 21])
    plt.show()

# 1回目のプロット (精度)
plot_data(N=-1, ylabel='Accuracy[%]', title='')

# 2回目のプロット (実行時間)
plot_data(N=1, ylabel='Execution time[s]', title='')
```
